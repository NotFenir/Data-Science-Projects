## Cleaning data

```python
df['date'] = pd.to_datetime(df['date'])
```

Ordering age_group

```python
df['age_group'].unique()
```

```python
order = ['13-18', '19-25', '26-35', '36-45', '46-60', '60+']
df['age_group'] = pd.Categorical(df['age_group'], categories=order, ordered=True)
```
