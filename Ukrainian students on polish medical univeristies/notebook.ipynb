## EDA

```python
df.columns
```

```python
df_plot = (
    df
    .groupby('wojewodztwo')[['studenci_suma', 'studenci_pl', 'studenci_ua']]
    .sum()
    .reset_index()
    .sort_values(by='studenci_suma', ascending=False)
)
```

```python
sns.barplot(df_plot, y='wojewodztwo', x='studenci_suma', palette='viridis')
plt.grid(axis='x')
plt.tight_layout()
```

```python
sns.barplot(df_plot.sort_values(by='studenci_ua', ascending=False), y='wojewodztwo', x='studenci_ua', palette='viridis')
plt.grid(axis='x')
plt.tight_layout()
```

```python
sns.barplot(df_plot.sort_values(by='studenci_pl', ascending=False), y='wojewodztwo', x='studenci_pl', palette='viridis')
plt.grid(axis='x')
plt.tight_layout()
```

```python
df_mean = (
    df
    .groupby('wojewodztwo')[['studenci_suma', 'studenci_pl', 'studenci_ua']]
    .mean()
    .reset_index()
    .sort_values(by='studenci_suma', ascending=False)
)


sns.barplot(
    data=(
        df_mean
        .assign(ua_suma_ratio=lambda x: x['studenci_ua'] / x['studenci_suma'])
        .sort_values(by='ua_suma_ratio', ascending=False)
    ), 
    x='ua_suma_ratio', 
    y='wojewodztwo', 
    hue='wojewodztwo',
    palette='viridis',
)
plt.grid(axis='x')
plt.tight_layout()
```
